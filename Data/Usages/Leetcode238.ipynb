Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in O(n) time and without using the division operation.

 

Example 1:

Input: nums = [1,2,3,4]
Output: [24,12,8,6]

Example 2:

Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]

 

Constraints:

    2 <= nums.length <= 105
    -30 <= nums[i] <= 30
    The input is generated such that answer[i] is guaranteed to fit in a 32-bit integer.


```python
class Solution:
    def productExceptSelf(self, nums: list[int]) -> list[int]:
        
        n = len(nums)
        Left_product = [1] * n
        Right_prod   = [1] * n
        res =          [0] * n
        # Calculate products of all elements to the left
        for i in range(1,n) :
            Left_product[i] = Left_product[i-1] * nums[i-1] 
            # print(Left_product[i])
            

        # Calculate products of all elements to the right
        for i in range(n-2,-1,-1) :
            Right_prod [i] = Right_prod [i+1] * nums[i+1]
            # print(Right_prod[i+1],nums[i+1],n,i)
            # print(Right_prod[i])
        
        for i in range(n) :
                    # print(Left_product[i],Right_prod[i])
                    res[i] = Left_product[i] *Right_prod[i]
        return res
sol =Solution()
res =sol.productExceptSelf([1,2,3,4])
print(res)                
```
