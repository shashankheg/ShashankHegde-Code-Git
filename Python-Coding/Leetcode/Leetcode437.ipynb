Given the root of a binary tree and an integer targetSum, return the number of paths where the sum of the values along the path equals targetSum.

The path does not need to start or end at the root or a leaf, but it must go downwards (i.e., traveling only from parent nodes to child nodes).

Input: root = [10,5,-3,3,2,null,11,3,-2,null,1], targetSum = 8
Output: 3
Explanation: The paths that sum to 8 are shown.

Example 2:

Input: root = [5,4,8,11,null,13,4,7,2,null,null,5,1], targetSum = 22
Output: 3


```python
from collections import deque

class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right

class Solution:
    def pathSum(self, root: TreeNode, targetSum: int) -> int:
        self.count = 0
        
        # DFS from current node
        def dfs(node, totalsum):
            if not node:
                return
            totalsum += node.val
            if totalsum == targetSum:
                self.count += 1
            dfs(node.left, totalsum)
            dfs(node.right, totalsum)
        
        # DFS starting from every node
        def solve(node):
            if not node:
                return
            dfs(node, 0)
            solve(node.left)
            solve(node.right)
        
        solve(root)
        return self.count

# Tree builder from level-order list
def build_tree(arr):
    if not arr:
        return None
    root = TreeNode(arr[0])
    queue = deque([root])
    i = 1
    while queue and i < len(arr):
        node = queue.popleft()
        if i < len(arr) and arr[i] is not None:
            node.left = TreeNode(arr[i])
            queue.append(node.left)
        i += 1
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            queue.append(node.right)
        i += 1
    return root

# Example usage
arr = [10,5,-3,3,2,None,11,3,-2,None,1]
root = build_tree(arr)

sol = Solution()
res = sol.pathSum(root, targetSum=22)
print(res)  # Output: 3
```
